```python
import pandas as pd
from collections import Counter

all_matches = pd.read_csv('2015_2019_All_Match_Results.csv', encoding  = 'cp949')
all_matches

club_tourney = []
#동호인 제외
for i in range(len(all_matches)):
    if '동호인' in all_matches['Tournament'][i]:
        club_tourney.append(i)
    
    
#선수권 대회
pro_matches = all_matches.drop(club_tourney, axis = 0)
match_count = Counter(pro_matches['Player']).most_common()

#승리한 경기
won_matches = pro_matches[pro_matches['Result'] == '승'].reset_index(drop=True)

#남자부 경기
won_matches_m = won_matches[won_matches['Gender'] == '남'].reset_index(drop=True)
count = Counter(won_matches_m['Player']).most_common()


#남자부 선수
players_m = pro_matches['Player'][pro_matches['Gender'] == '남'].drop_duplicates().sort_values().reset_index(drop =True)

players_m
```

```python
match_dic={}

for data in match_count:
    match_dic[data[0]] = data[1]
```

```python
win_dic = {}

for data in count:
    win_dic[data[0]] = data[1]
```

```python
per_dic = {}

for key in win_dic:
    results = []
    
    won = win_dic[key]
    results.append(won)
    matches = match_dic[key]
    results.append(matches)
    percent = round(float(won)/float(matches)*100, 2)
    results.append(percent)
    
    per_dic[key] = results
    
```

```python
result_df = pd.DataFrame().from_dict(per_dic, orient='index')
result_df.columns = ['Won', 'Matches', "Win_%"]
result_df['Starting Elo'] = 0
#result_df['Weighted Starting El']
result_df['Category'] = '일반부'
result_df.sort_values('Win_%', ascending = False)


for i in range(len(result_df)):
    win_per = result_df['Win_%']
    if win_per[i] > 80:
        result_df['Starting Elo'][i] = 1800
    elif 80 >= win_per[i] > 70:    
        result_df['Starting Elo'][i] = 1600
    elif 70 >= win_per[i] > 60:
        result_df['Starting Elo'][i] = 1400
    elif 60 >= win_per[i] > 50:
        result_df['Starting Elo'][i] = 1200
    else:
        result_df['Starting Elo'][i] = 1000

result_df = result_df.reset_index()
result_df
```

```python
def expected_result(elo_a, elo_b):
    """
    https://en.wikipedia.org/wiki/Elo_rating_system#Mathematical_details
    """
    expect_a = 1.0/(1+10**((elo_b - elo_a)/elo_width))
    return expect_a
```

```python
def update_elo(winner_elo, loser_elo):
    """
    https://en.wikipedia.org/wiki/Elo_rating_system#Mathematical_details
    """
    expected_win = expected_result(winner_elo, loser_elo)
    change_in_elo = k_factor * (1-expected_win)
    winner_elo += change_in_elo
    loser_elo -= change_in_elo
    return winner_elo, loser_elo
```

```python
elo_width = 400
k_factor = 20

for i in range(len(won_matches_m)):
    try:
        winner = won_matches_m['Player'].iloc[i]
        loser = won_matches_m['Oppoenent'].iloc[i]

        winner_elo = elo_dic[str(winner)]
        loser_elo = elo_dic[str(loser)]

        elo_dic[str(winner)], elo_dic[str(loser)] = update_elo(winner_elo, loser_elo)
    except KeyError:
        pass
```

```python
elo_dic = dict(sorted(elo_dic.items(), key=lambda item: item[1], reverse = True))
```

```python
ranking_df = pd.DataFrame.from_dict(elo_dic, orient = 'index')
ranking_df = ranking_df.reset_index()

ranking_df = ranking_df.rename(columns = {'index':'Name', 0: 'Rating'} )
```
