```python
import pandas as pd
import re
```

```python
com_df = pd.read_csv("Player_Results_CSV_Final.csv")

for i in range(2,7):
    file = "Player_Results_CSV_Final_"+str(i)+".csv"
    df = pd.read_csv(file)
    com_df = com_df.append(df)
    

```

```python
no_dup = com_df.drop_duplicates()
no_dup = no_dup.reset_index(drop = True)
```

```python
for i in range(len(no_dup)):
    if no_dup.Year[i] == 'Year':
        print(i)
```

```python
#no_dup[no_dup['Year'] == "Year"]
no_dup = no_dup.drop([4658], axis = 0)
DF = no_dup.reset_index(drop=True)

DF['Year'] = DF['Year'].astype('int64')
```

```python
DF = DF.rename({'Team': "Player Team", 'Team.1': 'Opponent Team'}, axis = 1)
DF
```

```python
from tqdm import tqdm

last_5yrs = pd.DataFrame()

for idx in tqdm(range(len(DF))):
    if DF.Year.iloc[idx] >= 2015:
        last_5yrs = last_5yrs.append(DF.iloc[idx])
```

```python
last_5yrs = last_5yrs.reset_index(drop=True)
last_5yrs['Gender'] = 'human'
last_5yrs['Category'] = 'sth'

#keyword = re.compile('남자')
#result = re.search(keyword, div)

for i in range(len(last_5yrs)):
    div = last_5yrs.Division[i]
    #keyword = re.compile('남자')
    #result = re.search(keyword, div)
    splits = div.split('자')
    gender = splits[0]
    category = splits[1]
    last_5yrs['Gender'][i] = gender
    last_5yrs['Category'][i] = category
    #print(splits[0])

last_5yrs
```

```python
count =0
wo = []

for i in range(len(last_5yrs)):
    score = last_5yrs['Player Game Scores'][i]
    if score == str([0,0,0,0,0]):
        count+=1
        wo.append(i)


last_5yrs = last_5yrs.drop(wo, axis = 0)
last_5yrs

```

```python
last_5yrs.to_excel("2015_2019_All_Match_Results.xlsx", index = None)
last_5yrs.to_csv("2015_2019_All_Match_Results.csv", index = None)
```

## Player Data

```python
all_data = pd.read_csv("2015_2019_All_Match_Results.csv", encoding = 'cp949')
```

```python
players = all_data.Player.drop_duplicates().reset_index(drop = True)
players  
```

```python
dic = {}

for player in tqdm(players):
    player_result = all_data[all_data.Player == player].reset_index(drop = True)
    match_count = len(player_result)
    wins = len(player_result[player_result.Result == '승'])
    loss = len(player_result[player_result.Result == '패'])
    NAN = len(player_result[player_result.Result == 'nan'])
    gender = player_result.Gender[0]
    
    win_per = round(wins/(match_count-NAN)*100, 2)
    ls = [gender, match_count, wins, loss, NAN, win_per]
    dic[player] = ls
    
```

```python
dfdf = pd.DataFrame().from_dict(dic).T
dfdf.columns = ['Gender','Total Matches', 'Win', 'Loss', 'N/A', "Win %"]
dfdf
```

```python
dfdf.to_excel("Player_Win_Percentage_Summary_2015_2019.xlsx")
dfdf.to_csv("Player_Win_Percentage_Summary_2015_2019.csv")
```

## Total Sets

```python
keyword1 = re.compile("실업")
keyword2 = re.compile("동호인")
keyword3 =re.compile("초등")

best_of_5 = all_data

for x in range(len(all_data)):
    tourney = all_data.Tournament[x]
    division =all_data.Division[x]
    
    result1 = re.search(keyword1,tourney)
    result2 = re.search(keyword2, tourney)
    result3 = re.search(keyword3, division)
    if (result1 or result2):
        best_of_5 = best_of_5.drop(x, axis=0)
        #print(tourney)
        
    if result3:
        best_of_5 = best_of_5.drop(x, axis=0)
        #print(division)
best_of_5
```

```python
dic2={}
players = best_of_5.Player.drop_duplicates().reset_index(drop = True)
 

for player in tqdm(players):
    player_result = best_of_5[best_of_5.Player == player].reset_index(drop = True)
    matches_won = player_result[player_result['Result'] == '승']
    matches_lost = player_result[player_result['Result'] == '패']
    
    matches_won_sets_won = matches_won['Player Games Won'].astype('int64').sum()
    matches_won_sets_lost = matches_won['Opponent Games Won'].astype('int64').sum()
    number_matches_won = len(matches_won)
    
    
    matches_lost_sets_won = matches_lost['Player Games Won'].astype('int64').sum()
    matches_lost_sets_lost = matches_lost['Opponent Games Won'].astype('int64').sum()
    number_matches_lost = len(matches_lost)
    
    sets_won = matches_won_sets_won+matches_lost_sets_won
    sets_lost = matches_won_sets_lost + matches_lost_sets_lost
    
    total_sets = sets_won+sets_lost
    total_matches = len(player_result)
    
    gender = player_result.Gender[0]
    
    set_ratio = round(sets_won/sets_lost, 2)
    #sets_won_per_match_won = round(matches_won_sets_won/number_matches_won, 2)
    sets_lost_per_match_won = round(matches_won_sets_lost/number_matches_won, 2)
    
    sets_won_per_match_lost = round(matches_lost_sets_won/number_matches_lost, 2)
    #sets_lost_per_match_lost = round(matches_lost_sets_lost/number_matches_lost, 2)
    win_percent = round(number_matches_won/total_matches, 2)
    
    
    ls2 = [gender, total_matches, total_sets, win_percent, number_matches_won, matches_won_sets_lost, number_matches_lost, matches_lost_sets_won,
           sets_won, sets_lost, sets_lost_per_match_won, sets_won_per_match_lost, set_ratio]
    dic2[player] = ls2
    
```

```python
df2 = pd.DataFrame().from_dict(dic2).T
df2.columns = ["Gender", "Total Matches", "Total Sets", "Win %", "Matches Won", "Sets lost for Won Matches", "Matches Lost", "Sets won for Lost Matches", 
               "Total Sets Won", "Total Sets Lost","Average Sets Lost per Won Matches", "Average Sets Won for Lost Matches", "Sets Won: Sets Lost"]
df2.to_excel("Player Match Set Data.xlsx")

df2
```
